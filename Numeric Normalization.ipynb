#### Normalize the dataset

```python
#Converting the dataset into an array and then normalizing using Sklearn
from sklearn import preprocessing    #we will utilize the sklearn library
import numpy as np

data_array = np.array(data['X'])        #Step 1: convert pandas series data into numpy array

norm_data = preprocessing.normalize([data_array])  #step 2: normalize
print(norm_data)
```

```python
import matplotlib.pyplot as plt
norm_data.plot(kind = 'bar')
```

```python
#When dealing with more than 1 column in the dataset
data1 = pd.read_csv("mpg_norm.csv")
# c_mpg = pd.DataFrame(car_mpg)
print(data1)
```

```python
data1
```

```python
c_mpg = data1.fillna(0)
c_mpg
```

```python
c_mpg.describe()
```

```python
d = preprocessing.normalize(c_mpg)
colName = c_mpg.columns
scaled_df = pd.DataFrame(d, columns=colName)
scaled_df
```

#### Normalize with a Scaler

```python
c_mpg
```

```python
#Using the Min_max_scaler
scaler = preprocessing.MinMaxScaler()
colName = c_mpg.columns
d = scaler.fit_transform(c_mpg)
scaled_df = pd.DataFrame(d, columns=colName)
scaled_df.head()
```

```python
import matplotlib.pyplot as plt
scaled_df.plot(kind='bar')
```

```python
# copy the data
df_z_scaled = data1.copy()

# apply normalization techniques
for column in df_z_scaled.columns:
    df_z_scaled[column] = (df_z_scaled[column] -
                           df_z_scaled[column].mean()) / df_z_scaled[column].std()

# view normalized data
display(df_z_scaled)
```

```python
import matplotlib.pyplot as plt
df_z_scaled.plot(kind='bar')
```
