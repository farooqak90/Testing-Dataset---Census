## Data Transformation
Data transformation deals with normalizing also known as scaling data , handling skewness and aggregation of attributes.

involves converting variables into a format that can be easily analyzed.

This includes encoding categorical variables, scaling numerical variables, and reducing dimensionality.


## Data Normalization and Scaling
> Data normalization and scaling are important steps in data preprocessing before modeling.
## Uses:
> data can be easily compared and contrasted with other data sets
> These techniques help to bring all the features to a similar scale/ range , which is essential for certain machine learning algorithms to work effectively.

>particularly useful for models that require inputs to be on the same scale, such as K-nearest neighbors and artificial neural networks.
> Scaling is useful for models that require features to be on the same scale, such as support vector machines and linear regression.

> In this Session, we will discuss how to normalize and scale data using pandas library in Python.


## Normalization:
> reshaping to specific format
>Normalization is the process of transforming the data to a common scale.
>
>objective : to  rescale the features to a range of 0 to 1.

>This makes it easier to compare the data as it eliminates the effects of the scale on the analysis.

>Pandas provides a convenient way to normalize data using the MinMaxScaler class from the sklearn.preprocessing module.


## scaling:
> increasing or decreasing the size of data.

```python
# example for MinMaxScaler usage to normalize the data.
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# read the data from a csv file
# data = pd.read_csv('data.csv')

# Create a sample dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})
```

```python
# create a MinMaxScaler object
scaler = MinMaxScaler()
```

```python
# fit and transform the data using fit_transform method
normalized_data = scaler.fit_transform(df)
```

```python
# create a new DataFrame with the normalized data
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)
print("Raw Data")
print(df)
```

```python
print("\nNormalized Data")
print(normalized_df)
```

## Scaling
Scaling is the process of transforming the data to a new scale "without changing the shape of the data distribution".
Scaling is useful for data that has a wider range of values or different units of measurement.
Pandas provides a way to scale data using the StandardScaler class from the sklearn.preprocessing module.
prevents any one feature from dominating the others

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# read the data from a csv file
# data = pd.read_csv('data.csv')

# Create a sample dataframe
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})
```

```python
# create a StandardScaler object
scaler = StandardScaler()
```

```python
# fit and transform the data
scaled_data = scaler.fit_transform(df)
```

```python
# create a new DataFrame with the scaled data and columns
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("Raw Data")
print(df)
print("\nNormalized Data")
print(scaled_df)
```

```python
# skewness along the index axis
iris.skew(axis = 0, skipna = True)
```

If skewness value lies above +1 or below -1, data is highly skewed.
If it lies between +0.5 to -0.5, it is moderately skewed.
If the value is 0, then the data is symmetric

```python
from matplotlib import pyplot as plt

# Compute skewness and plot
skewness = iris.skew()
skewness.plot(kind='bar')
plt.show()
```
